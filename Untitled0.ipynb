```python
############################################
# 1. ЗАГРУЗКА ФАЙЛОВ (train.csv, test.csv)
############################################

from google.colab import files
uploaded = files.upload()
# Выберите train.csv и test.csv

############################################
# 2. ПРОСМОТР СОДЕРЖИМОГО /content
############################################

import os

for dirpath, _, filenames in os.walk('/content/'):
    for filename in filenames:
        print(os.path.join(dirpath, filename))

!ls /content

############################################
# 3. ЧТЕНИЕ train.csv "по-строчно" (пример)
############################################

print("\nПервые 4 строки train.csv (через open):")
with open("/content/train.csv", "r") as f:
    for i, line in enumerate(f):
        if i < 4:
            print(line.strip())
        else:
            break

############################################
# 4. ПРЕОБРАЗОВАНИЕ В DATAFRAME (pandas)
############################################

import pandas as pd

train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test.csv")

print("\n>>> train.head() <<<")
print(train.head())

print("\n>>> train.info() <<<")
train.info()

############################################
# 5. ОБРАБОТКА ПРОПУСКОВ И УДАЛЕНИЕ 'CABIN'
############################################

# Заполняем Age и Embarked в train
train["Age"].fillna(train["Age"].mean(), inplace=True)
train["Embarked"].fillna(train["Embarked"].mode()[0], inplace=True)

# Удаляем Cabin (много пропусков), а также позже Name, Ticket
train.drop(columns=["Cabin"], inplace=True, errors="ignore")

print("\n>>> Пропуски (train.isnull().sum()) после fillna и drop('Cabin'): <<<")
print(train.isnull().sum())

############################################
# 6. ВИЗУАЛИЗАЦИЯ (countplot)
############################################

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Pclass", hue="Survived", data=train)
plt.show()

############################################
# 7. ПРЕОБРАЗОВАНИЕ КАТЕГОРИАЛЬНЫХ ПРИЗНАКОВ
############################################

train = pd.get_dummies(train, columns=["Sex", "Embarked", "Pclass"], drop_first=True)

# Удаляем нечисловые/ненужные столбцы
train.drop(columns=["Name", "Ticket", "PassengerId"], inplace=True, errors='ignore')

print("\n>>> train.columns после get_dummies и удаления строковых: <<<")
print(train.columns)

# Делим на X_train, y_train
y_train = train["Survived"]
X_train = train.drop(columns=["Survived"])

############################################
# 8. ОБУЧЕНИЕ ЛОГИСТИЧЕСКОЙ РЕГРЕССИИ
############################################

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_train)
cm = confusion_matrix(y_train, y_pred)
acc = accuracy_score(y_train, y_pred)

print("\n>>> Матрица ошибок (train) <<<")
print(cm)
print("\n>>> Accuracy (train) <<<")
print(acc)

############################################
# 9. АНАЛОГИЧНАЯ ОБРАБОТКА test.csv
############################################

# Заполним Age, Fare и Embarked
# (в test.csv часто есть пропуск в Fare, поэтому явно делаем fillna)
test["Age"].fillna(test["Age"].mean(), inplace=True)
test["Fare"].fillna(test["Fare"].mean(), inplace=True)
test["Embarked"].fillna(test["Embarked"].mode()[0], inplace=True)

# Удаляем Cabin
test.drop(columns=["Cabin"], inplace=True, errors="ignore")

# Сохраняем PassengerId для итогового результата
test_ids = test["PassengerId"]

# Преобразуем test: get_dummies
test = pd.get_dummies(test, columns=["Sex", "Embarked", "Pclass"], drop_first=True)

# Удаляем нечисловые/ненужные
test.drop(columns=["Name", "Ticket", "PassengerId"], inplace=True, errors='ignore')

# Проверяем, есть ли пропуски
print("\n>>> test.isnull().sum() после fillna <<<")
print(test.isnull().sum())

# Приводим test к тем же столбцам, что и X_train
for col in X_train.columns:
    if col not in test.columns:
        test[col] = 0
for col in test.columns:
    if col not in X_train.columns:
        test.drop(columns=[col], inplace=True)

test = test[X_train.columns]

############################################
# 10. ПОЛУЧЕНИЕ ПРОГНОЗА НА test.csv
############################################

y_pred_test = model.predict(test)

df_result = pd.DataFrame({
    "PassengerId": test_ids,
    "Survived": y_pred_test
})

print("\n>>> Пару строк результата <<<")
print(df_result.head())

df_result.to_csv("result.csv", index=False)

############################################
# 11. СКАЧАТЬ result.csv
############################################

files.download("/content/result.csv")

print("\nГотово! Проверяйте result.csv и убедитесь, что нет NaN при predict().")
```
